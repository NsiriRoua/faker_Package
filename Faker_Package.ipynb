**Why not we create a customized profile**

```python
from faker import Faker
import json
import numpy as np
fake=Faker()
def create_profiles(x):

  profiles_data={}
  for i in range(0,x):
    profiles_data[i]={} 
    profiles_data[i]['name']=fake.name()
    profiles_data[i]['address']=fake.address()
    profiles_data[i]['city']=fake.city()
    profiles_data[i]['color']=fake.city()
    profiles_data[i]['closeness (1-5)'] = np.random.randint(0,5)
    return profiles_data

profiles=create_profiles(8)
profiles_df=pd.DataFrame(profiles_df)
profiles_df.head()
```
